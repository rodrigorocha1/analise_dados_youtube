```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
import pyhive
from pyhive import hive
```

```python
spark =SparkSession.builder.appName('CARGA').getOrCreate()
spark
```

```python
df_comentarios = spark.read \
    .parquet('hdfs://localhost:9000/projeto/datalake_youtube/comentarios.parquet')
df_comentarios.show(truncate=False)
```

```python
df_comentarios.groupBy('ASSUNTO').agg(F.count('ASSUNTO')).show()
```

```python
lista_assunto = [
        'assunto_cities_skylines',
        'assunto_cities_skylines_2',
        'assunto_power_bi',
        'assunto_python_and_dados'
]
```

```python
for assunto in lista_assunto:
    df_comentarios.select('ID_COMENTARIO', 'TEXTO_COMENTARIO', 'ASSUNTO', 'ID_CANAL', 'ID_VIDEO') \
        .filter(df_comentarios.ASSUNTO == assunto) \
        .write \
        .parquet(f'../data/projetos_youtube_v2/teste/comentarios_{assunto}.parquet')
```
