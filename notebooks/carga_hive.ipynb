- Comentários

```python
df_comentarios = spark.read \
    .parquet('hdfs://localhost:9000/projeto/datalake_youtube/comentarios.parquet')
df_comentarios.show(truncate=False)
```

```python
colunas = df_comentarios.columns
```

```python
for coluna in colunas:  
    df_comentarios.groupBy(coluna).count().orderBy('count', ascending=False).show()
```

```python
df_comentarios.select('ID_VIDEO').distinct().collect()
```

```python
lista_videos = [ linha['ID_VIDEO'] for linha in df_comentarios.select('ID_VIDEO').distinct().collect()]
lista_videos
```

```python
len(df_comentarios.select('ID_VIDEO').distinct().collect())
```

```python
for video in lista_videos:
    df_comentarios \
        .filter(df_comentarios.ID_VIDEO == video) \
        .write \
        .parquet(f'hdfs://localhost:9000/projeto/datalake_youtube/particao_por_campo/comentarios/comentarios_{video}.parquet')
```
